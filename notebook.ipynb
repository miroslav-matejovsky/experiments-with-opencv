```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

original = cv2.imread("coffees.jpg")

image = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)

```

```python
# Threshold for darkest places
_, threshold_dark = cv2.threshold(image, 30, 255, cv2.THRESH_BINARY_INV)

# Threshold for lightest places
_, threshold_light = cv2.threshold(image, 180, 255, cv2.THRESH_BINARY)

# Display the original image and the thresholded images
titles = ['Original Image', 'Darkest places', 'Lightest places']
images = [image, threshold_dark, threshold_light]

for i in range(3):
    plt.subplot(1, 3, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.axis('off')

plt.show()
```

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Define range for the first color filter
lower_range1 = np.array([0, 0, 0])
upper_range1 = np.array([150, 150, 150])
mask1 = cv2.inRange(image, lower_range1, upper_range1)
filtered_image1 = cv2.bitwise_and(image, image, mask=mask1)

# Define range for the second color filter
lower_range2 = np.array([150, 150, 150])
upper_range2 = np.array([255, 255, 255])
mask2 = cv2.inRange(image, lower_range2, upper_range2)
filtered_image2 = cv2.bitwise_and(image, image, mask=mask2)

# Combine the two filters to create a black and white image
bw_image = cv2.add(filtered_image1, filtered_image2)

# Display the original image, the filtered images, and the black and white image
titles = ['Original Image', 'First Color Filter', 'Second Color Filter', 'Black and White Image']
images = [image, filtered_image1, filtered_image2, bw_image]

for i in range(4):
    plt.subplot(1, 4, i+1), plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')

plt.show()
```
